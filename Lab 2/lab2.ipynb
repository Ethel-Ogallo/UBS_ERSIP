```python
# computing a histogram
import numpy as np
import time
import rasterio
```

```python
img = rasterio.open(r"C:\Users\Ethel Ogallo\Documents\CDE\UBS\ERSIP\UBS_ERSIP\Lab 1\spot2005_pansharpened_utm33_subset_NIRRGMIR.tif")
input =img.read(1)
```

```python
# naive approach 
start = time.process_time()

hist=np.zeros(256,dtype=int)
values=input.flatten()
for v in range(len(hist)):
    for i in values:
        if i==v:
            hist[v]+=1

end = time.process_time()
print("Naive approach time:", end - start)
```

```python
# efficient 
start = time.process_time()

hist=np.zeros(256,dtype=int)
values=input.flatten()
for i in values:
    hist[i]+=1

end = time.process_time()
print("Efficient approach time:", end - start)
```

```python
# efficient 2
start = time.process_time()

hist=np.histogram(input,256,range=(0,255))[0]

end = time.process_time()
print("Efficient 2 approach time:", end - start)
```

```python
# plot histogram
import matplotlib.pyplot as plt

plt.plot(hist)
```

```python
plt.hist(input.flatten(),bins=256,range=[0,255])
```

```python
# Automatic contrast stretching#
nz=np.nonzero(hist)[0]
min,max=nz[0],nz[-1]
print("min:",min," max:",max)  
```

```python
# cumulative histogram
chist=hist
for i in range(1,len(hist)):
    chist[i]=chist[i-1]+hist[i]

plt.plot(chist)
```
